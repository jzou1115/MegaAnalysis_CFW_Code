# size of confidence interval

```python
ci["size"] = ci["to.bp"] - ci["from.bp"]
```

```python
for q in set(ci["quantile"]):
    sub = ci[ci["quantile"]==q]
    sns.distplot(sub["size"], kde=False, label=q)
plt.legend()
```

```python
for q in [0.75, 0.9, 0.95]:
    sub = ci[ci["quantile"]==q]
    sns.distplot(sub["size"]/1000000, kde=False, label=q)
plt.legend(title="Quantile")
plt.xlabel("Size of confidence interval (MB)")
plt.ylabel("Number of QTLs")
plt.savefig("/u/home/j/jzou1115/project-zarlab/CFW/confidence_intervals/output/confidence_intervals/good/CI_pruned_sizes.pdf")
```

```python
phenotype="tibia"
chrm=2
bp=153762928
start=153762928.0
end=153765764.0
snp_logp = 5.135886
```

```python
stats=pd.read_table("/u/home/j/jzou1115/project-zarlab/CFW/MegaAnalysis_CFW_2019_Summary_Statistics/COMBINED/tibia.combined.assoc.txt.gz")
stats.head()
```

```python
locus = stats[stats["chr"]==2]
locus.shape
```

```python
window=5000000
locus = locus[(locus["ps"] > bp-(window/2)) & (locus["ps"]< bp + (window/2))] 
```

```python
locus.shape
```

```python
ci = locus[(locus["ps"] > start) & (locus["ps"] < end)] 
ci.shape
```

```python
locus = locus.sort_values(by="ps")
locus["logp"] = -1*np.log10(locus["p_wald"])
plt.scatter(locus["ps"], locus["logp"])
plt.scatter([bp], [snp_logp], c="red")
plt.plot([start, start], [0,5], c="black")
plt.plot([end, end], [0,5], c="black")
plt.xlabel("Genomic position (bp)")
plt.ylabel("-log10(p-value)")
locus_name=phenotype+"_"+str(chrm)+"_"+str(bp)
plt.title(locus_name)
plt.savefig("/u/home/j/jzou1115/project-zarlab/CFW/confidence_intervals/output/confidence_intervals/good/"+locus_name+".pdf")
```

```python
ci_bed = pd.read_table("/u/home/j/jzou1115/project-zarlab/CFW/confidence_intervals/output/confidence_intervals/good/CI_pruned_q95_genes.txt", header=None)
ci_bed.shape
```

```python
ci_bed.columns = ["chr", "bp.from", "bp.to", "qtl", "chr", "gene.from", "gene.to", "gene"]
ci_bed.head()
```

```python
gene_count = {}
gene_count["qtl"] = []
gene_count["count"] = []
gene_count["genes"] = []
for q in set(ci_bed["qtl"]):
    sub = ci_bed[ci_bed["qtl"] == q]
    gene_count["qtl"].append(q)
    gene_count["count"].append(sub.shape[0])
    genes = ",".join(sub["gene"])
    gene_count["genes"].append(genes)
gene_count = pd.DataFrame(gene_count)    
```
