# combined

```python
qtls = pd.read_csv("/u/home/j/jzou1115/project-zarlab/CFW/MegaAnalysis_CFW_2019_Summary_Statistics/Flint_July14/CombinedPeaks160620.txt", sep="\t")
qtls.head()
```

```python
len(".assoc.90info.txt.gz")
```

```python
pheno = [p[:-20] for p in qtls["phenotype"]]
qtls["phenotype"] = pheno
set(pheno)
```

```python
fc_phenos = ['fc.baseline',
 'fc.context',
 'fc.context.corr',
 'fc.cue',
 'fc.cue.baseline',
 'fc.cue.corr',
 'fc.uncond.freeze',
 'fc.uncond.freeze.corr']
print(len(fc_phenos))
```

```python
qtls_fc = []
for p in fc_phenos:
    qtls_fc.append(qtls[qtls["phenotype"]==p])
qtls_fc = pd.concat(qtls_fc, ignore_index=True)
qtls_fc.describe() #no qtls have combined.logp> 5.787229
```

```python
threshold=5.787229 
sig=qtls[qtls["combined.logp"]>threshold]
sig.shape
```

```python
#filter qtls that do not have info score > 90 
sig["info90"] = False
for chrm in set(sig["chr"]):
    sig_chrm = sig[sig["chr"]==chrm]
    geno = pd.read_table("/u/home/j/jzou1115/project-zarlab/CFW/MegaAnalysis_CFW_2019/genotypes/oxinfo90/combined.chr"+str(chrm)+".oxinfo90.annot", header=None)
    markers = set(list(geno[0]))
    for m in sig_chrm["marker"]:
        if str(m) in markers:
            sig["info90"] = True
```

```python
markers90 = {}
sig["info90"] = False
for i in sig.index:
    chrm = sig.loc[i, "chr"]
    if chrm not in markers90.keys():
        geno = pd.read_table("/u/home/j/jzou1115/project-zarlab/CFW/MegaAnalysis_CFW_2019/genotypes/oxinfo90/combined.chr"+str(chrm)+".oxinfo90.annot", header=None)
        markers90[chrm] = set(geno[0])

    markers = markers90[chrm]
    m = sig.loc[i, "marker"]

    if m in markers:
        sig.loc[i, "info90"] = True
```

```python
sig_info90 = sig[sig["info90"]==True]
sig_info90.shape
```

```python
# make qtl input file
#columns:
#phenotype chr bp logP fdr.adj threshold start end dist
out = pd.DataFrame({})
out["phenotype"]= sig_info90["phenotype"]
out["chr"] = sig_info90["chr"]
out["bp"] = sig_info90["pos"]
out["logP"] = sig_info90["combined.logp"]
out["fdr.adj"] = sig_info90["combined.logp"]
out["threshold"] = 0.05
out["start"] = sig_info90["pos"] - 1000000
out["end"] = sig_info90["pos"] + 1000000
out["dist"] = 2000000
```

```python
out.to_csv("/u/home/j/jzou1115/project-zarlab/CFW/confidence_intervals/data/qtls/qtls_decorrelated_threshold3.txt", header=True, index=False, sep="\t")
```

```python
out2 = open("/u/home/j/jzou1115/project-zarlab/CFW/MegaAnalysis_CFW_Code/confidence_interval_pipeline_pruned/data/qtls_decorrelated_info90_all.txt", "w")
for i in out.index:
    p=out.loc[i, "phenotype"]
    c = str(int(out.loc[i, "chr"]))
    bp = str(int(out.loc[i, "bp"]))
    
    out2.write(".".join([p, c, bp])+"\n")
out2.close()                        
```
